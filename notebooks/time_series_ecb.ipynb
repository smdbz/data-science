```python
import os
import pandas as pd

os.getcwd()
```

```python
files = os.listdir(r"data/time-series-ecb/")
files
```

```python
pd.read_csv("data/time-series-ecb/" + files[0])
```

```python
main_df = pd.DataFrame()

for file in files:
    current_file = pd.read_csv("data/time-series-ecb/" + file, parse_dates=["DATE"], index_col="DATE").drop('TIME PERIOD', axis=1)

    current_file.rename(columns={list(current_file)[0]:input(list(current_file)[0])}, inplace=True)

    if len(main_df)==0:
        main_df = current_file
    else:
        main_df = main_df.join(current_file, how='inner')
```
