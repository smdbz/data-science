```python
import kagglehub
import pandas as pd
```

```python
competition_handle = "house-prices-advanced-regression-techniques"
competition_url = "https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/overview"
competition_train = "train.csv"
competition_test = "test.csv"

data = kagglehub.competition_download(competition_handle)
```

```python
train_df = pd.read_csv(data + "\\" + "train.csv")
```

```python
# View the data
train_df
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=1)

fig.add_trace(go.Scatter(
    x = train_df.index,
    y= train_df['SalePrice'],
    mode="markers",
    name = "Sales Price"
))
```

```python
import scipy.stats as stats
import numpy as np

(theoretical_quantiles, sample_quantiles), (slope, intercept, r) = stats.probplot(train_df['SalePrice'], dist="norm")
```

```python
line_x = np.array([np.min(theoretical_quantiles), np.max(theoretical_quantiles)])
line_y = slope * line_x + intercept
```

```python
# 4. Build the figure using graph_objects
fig = go.Figure()

# Add the scatter plot for the quantiles
fig.add_trace(go.Scatter(
    x=theoretical_quantiles,
    y=sample_quantiles,
    mode='markers',
    name='Data Quantiles',
    marker=dict(
        color='royalblue',
        size=7,
        opacity=0.6,
        line=dict(width=1, color='DarkSlateGrey')
    )
))

# Add the reference line
fig.add_trace(go.Scatter(
    x=line_x,
    y=line_y,
    mode='lines',
    name='Perfect Normal Distribution',
    line=dict(color='firebrick', width=2, dash='dash')
))

# 5. Format the layout
fig.update_layout(
    title='Normal Q-Q Plot',
    xaxis_title='Theoretical Quantiles (Standard Normal)',
    yaxis_title='Sample Quantiles',
    template='plotly_white',
    width=800,
    height=600,
    hovermode='closest'
)

fig.show()
```
