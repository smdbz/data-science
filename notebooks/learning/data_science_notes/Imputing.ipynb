```python
class AnalysisResult:
    def __init__(self, score, status):
        self.score = score
        self.status = status

    def _repr_html_(self):
        # This will render as formatted HTML in the notebook
        color = "green" if self.status == "Pass" else "red"
        return f"""
        <div style="border: 1px solid #ddd; padding: 10px;">
            <h3>Analysis Result</h3>
            <p>Score: <strong>{self.score}</strong></p>
            <p>Status: <span style="color: {color};">{self.status}</span></p>
        </div>
        """

    def __repr__(self):
        # Fallback for standard Python terminals
        return f"AnalysisResult(score={self.score}, status='{self.status}')"
```

```python
import ipywidgets as widgets
from IPython.display import display

def launch_control_panel():
    slider = widgets.IntSlider(value=50, min=0, max=100, description='Threshold:')
    button = widgets.Button(description='Run Analysis')
    output = widgets.Output()

    def on_button_clicked(b):
        with output:
            output.clear_output()
            print(f"Running analysis with threshold: {slider.value}")

    button.on_click(on_button_clicked)

    # Display the UI
    display(widgets.VBox([slider, button, output]))
```

```python
launch_control_panel()
```
