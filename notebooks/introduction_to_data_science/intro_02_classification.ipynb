```python
import seaborn as sns
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
from itertools import count

# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
import pandas as pd
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "customer_churn_dataset-testing-master.csv"

# Load the latest version
df = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "mohamed444ashraf/customer-churn-data",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

df
```

```python
# Separate the target variable from everything else
X_train, y_train = df.select_dtypes("number"), df.Churn
```

```python
# Using 10,000 iter because I got a warning that the model hadn't converged with lower numbers.
log_r_model = LogisticRegression(max_iter=10000).fit(X_train, y_train)
```

```python
log_r_model.predict(X_train)
```

```python
y_pred = log_r_model.predict(X_train)

print("Prediction on training set:", y_pred)
print("Accuracy on training set:", log_r_model.score(X_train, y_pred))
```
