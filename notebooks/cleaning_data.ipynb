## Dealing with missing data

- MAR: Missing at Random - the good guys
    - Not coming from a systematic error
    - Low in proportion
    - Lack of/no apparent correlation to other variables in terms of missing data

    - Actions available after identification = Imputation and Dropping

- MNAR - Missing not at random - the bad guys
    - Systematic error involved
    - High proportion
    - Strong evidence of correlation

    - Actions available after identification = Imputation

- MCAR: Missing Completely at Random
    - A low version of MAR
