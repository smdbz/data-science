## Modulo Challenge

- In python to find the remainder when you divide a number by another we can use the modulo operator.
- The modulo operator will return the remainder 
- For example if we do 10%7 the value will be 3 as there is one 7 in 10 but 3 remains

```python
10%3
# modulo symbol is %
```

### Task 1
- Write a while loop that prints out the index if the index is completely divisible by any of these numbers [2,5,17]

```python
index = 1

while index <= 1000:
    if len([x for x in [2,5,17] if index % x ==0]) >0:
        print(index)
    index +=1
```
