## Mismatching lists challenge

## Instructions
- There are two lists: discounts and product prices. The discounts represents fractions of the year, where 4 elements denotes 4 quarters of the year. 
- Product prices has 12 elements - each denoting a product that is going on sale for each month.
- Write a for loop that will multiply each product price by its equivalent discount. For example, the first quarter of product_prices should be multipled by (1-0.25) ie (1 - discounts[0]). 
Then the second quarter of product prices (30, 1000, 105) should be multipled by 1-0.1 ie: (1-discounts[1]).

## Additional constraints

- The number of values in discounts will always be less than or equal to the number of values in product_prices
- Your are not to use **if** statements in your code
- AI assistance is not allowed

```python
def apply_discount(discount, price):
    return np.array(price) * 1-discount
```

```python
groups = int(len(product_prices)/len(discounts))
print(groups)


new_prices = [apply_discount(discounts[i], product_prices[i*groups:(i+1)*groups]) for i in range(len(discounts))]

flat_list = [x for xs in new_prices for x in xs]
flat_list
```
