```python
import pandas as pd, numpy as np, os, plotly.express as px, matplotlib.pyplot as plt

main_path = r"C:\Users\Xps\Documents\Tutor\Python\Time Series\Manipulating time series\ECB Data"
os.chdir(main_path)

files = [f for f in os.listdir(main_path) if os.path.isfile(fr'{main_path}\{f}')]
files
```

```python
main_df = pd.DataFrame()

for file in files:
    current_file = pd.read_csv(file, parse_dates=['DATE'], index_col='DATE').drop('TIME PERIOD', axis=1)

    current_file.rename(columns ={list(current_file.columns)[0]:input(list(current_file.columns)[0])} , inplace=True)

    if len(main_df) ==0:
        main_df = current_file
    else:
        main_df = main_df.join(current_file, how='inner')

```

## Plotting time series data

```python
main_df.plot()
```
