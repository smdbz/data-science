## Regular Expressions

```python
import re
```

```python
text = "This is a good day."

# Checking for an example

if re.search("good", text):
    print("Wonderful!")
else:
    print("Alas :(")
```

```python
re.search("good", text)
```

In addition to checking for conditionals, we can segment a string. The work that regex does here is called tokenizing, where the string is separated into substrings based on patterns. Tokenizing is a core activity in natural language processing.

The findall() and split() functions will parse the string for us and return chunks.

```python
text = "Amy works diligently. Amy gets good grades. Our student Amy is successful."

re.split("Amy", text)
```

```python
re.findall("Amy", text)
```

```python
re.search("Amy", text)
```

## Patterns and Character Classes

```python
grades = "ACAAAABCBCBAA"
```

```python

print(re.findall("AC", grades)) # All AC in the string
print(re.findall("[AC]", grades)) # All As or Cs in the string
print(re.findall("[A-C]", grades)) # All As through Cs in the string
print(re.findall("AB|C", grades)) # All As or Cs in the string

print(re.findall("[^A]", grades)) # Not A
print(re.findall("^[A]", grades)) # A at the start of the string
```

```python
def l2_dist(a, b):
    result = ((a - b) * (a - b)).sum()
    result = result ** 0.5
    return result
```
